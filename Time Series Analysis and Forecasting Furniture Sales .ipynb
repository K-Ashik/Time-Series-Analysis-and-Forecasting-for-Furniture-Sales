We can also visualize our data using a method called time-series decomposition that allows us to decompose our time series into three distinct components: trend, seasonality, and noise. </font>

```python

from pylab import rcParams


# Set the figure size
rcParams['figure.figsize'] = 18, 8

# Perform seasonal decomposition
decomposition = sm.tsa.seasonal_decompose(y, model='additive')

# Create the plot with customized colors
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(18, 8))

# Set background color for the figure
fig.patch.set_facecolor('lightgrey')  # Change 'lightgrey' to the color you prefer

# Plot the original time series
decomposition.observed.plot(ax=ax1, color='blue')
ax1.set_title('Observed', fontsize=16)
ax1.set_facecolor('white')  # Change 'white' to the color you prefer for the subplot background
ax1.grid(True)

# Plot the trend component
decomposition.trend.plot(ax=ax2, color='green')
ax2.set_title('Trend', fontsize=16)
ax2.set_facecolor('white')  # Change 'white' to the color you prefer for the subplot background
ax2.grid(True)

# Plot the seasonal component
decomposition.seasonal.plot(ax=ax3, color='red')
ax3.set_title('Seasonal', fontsize=16)
ax3.set_facecolor('white')  # Change 'white' to the color you prefer for the subplot background
ax3.grid(True)

# Plot the residual component
decomposition.resid.plot(ax=ax4, color='purple')
ax4.set_title('Residual', fontsize=16)
ax4.set_facecolor('white')  # Change 'white' to the color you prefer for the subplot background
ax4.grid(True)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

# <font face="Times New Roman" size="5">4. Time series forecasting with ARIMA </font> 

<font face="Times New Roman" size="3">We are going to apply one of the most commonly used method for time-series forecasting, known as ARIMA, which stands for Autoregressive Integrated Moving Average.

ARIMA models are denoted with the notation ARIMA(p, d, q). These three parameters account for seasonality, trend, and noise in data: </font>

```python


# Define p, d, and q parameters for SARIMA model
p = d = q = range(0, 2)
pdq = list(itertools.product(p, d, q))

# Define seasonal p, d, and q parameters for SARIMA model
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]

print('Examples of parameter combinations for Seasonal ARIMA...')
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[1]))
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[2]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[3]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[4]))

# Your SARIMA model fitting code here...
for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = sm.tsa.statespace.SARIMAX(y,
                                            order=param,
                                            seasonal_order=param_seasonal,
                                            enforce_stationarity=False,
                                            enforce_invertibility=False)
            results = mod.fit(maxiter=1000, method='powell', burn=10)  # Adjust burn-in period as needed
  # Adjusted parameters
            print('ARIMA{}x{}12 - AIC:{}'.format(param, param_seasonal, results.aic))
        except Exception as e:
            print(f"Failed for {param}, {param_seasonal}: {str(e)}")
            continue
```

<font face="Times New Roman" size="3"> This step is parameter Selection for our furniture’s sales ARIMA Time Series Model. Our goal here is to use a “grid search” to find the optimal set of parameters that yields the best performance for our model.
The above output suggests that SARIMAX(1, 1, 1)x(1, 1, 0, 12) yields the lowest AIC value of 283.366 and therefore we should consider this to be optimal option.

</font>

# <font face="Times New Roman" size="5">5. Fitting the ARIMA model </font>


```python
mod = sm.tsa.statespace.SARIMAX(y,
                                order=(1, 1, 1),
                                seasonal_order=(1, 1, 0, 12),
                                enforce_stationarity=False,
                                enforce_invertibility=False)
results = mod.fit(maxiter=1000, method='powell', burn=10)  # Adjust burn-in period as needed

print(results.summary().tables[1])
```

<font face="Times New Roman" size="3"> The above table consists coefficients from a ARIMA model. The table shows the coefficients of the autoregressive (AR) terms, the moving average (MA) terms, and the seasonal autoregressive (SAR) terms, as well as the residual variance. The coefficients are estimated using maximum likelihood estimation.

- AR.L1: This is the coefficient of the first lagged value of the time series. A positive coefficient indicates that there is a positive autocorrelation between the current value of the time series and the previous value. A negative coefficient indicates that there is a negative autocorrelation between the current value of the time series and the previous value.

- ma.L1: This is the coefficient of the first lagged error term. A positive coefficient indicates that there is a positive autocorrelation between the error terms. A negative coefficient indicates that there is a negative autocorrelation between the error terms.

- ar.S.L12: This is the coefficient of the first lagged squared error term. A positive coefficient indicates that there is a positive autocorrelation between the squared error terms. A negative coefficient indicates that there is a negative autocorrelation between the squared error terms.

- sigma2: This is the residual variance, which is the variance of the error terms. </font>

<font face="Times New Roman" size="3"> We should always run model diagnostics to investigate any unusual behavior. </font>

```python
# Get residuals
residuals = results.resid

# Plot residuals over time
plt.figure(figsize=(12, 6))
plt.plot(residuals)
plt.title('Residuals Plot')
plt.xlabel('Time')
plt.ylabel('Residuals')
plt.grid(True)
plt.show()





```

```python
# Plot histogram of residuals
plt.figure(figsize=(8, 6))
plt.hist(residuals, bins=30, density=True, alpha=0.7)
plt.title('Histogram of Residuals')
plt.xlabel('Residuals')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
# QQ plot of residuals
plt.figure(figsize=(8, 6))
sm.qqplot(residuals, line='q', ax=plt.gca())
plt.title('QQ Plot of Residuals')
plt.grid(True)
plt.show()
```

```python
# Autocorrelation function (ACF) plot of residuals
plt.figure(figsize=(8, 6))
sm.graphics.tsa.plot_acf(residuals, lags=40, ax=plt.gca())
plt.title('Autocorrelation Function (ACF) Plot of Residuals')
plt.xlabel('Lag')
plt.ylabel('ACF')
plt.grid(True)
plt.show()
```

<font face="Times New Roman" size="3">It is not perfect, however, our model diagnostics suggests that the model residuals are near normally distributed.</font>

# <font face="Times New Roman" size="5">6. Validating forecasts </font>

<font face="Times New Roman" size="3">To help us understand the accuracy of our forecasts, we compare predicted sales to real sales of the time series, and we set forecasts to start at 2017–01–01 to the end of the data. </font>

```python
pred = results.get_prediction(start=pd.to_datetime('2017-01-01'), dynamic=False)
pred_ci = pred.conf_int()
ax = y['2014':].plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='One-step ahead Forecast', alpha=.7, figsize=(14, 7))
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.2)
ax.set_xlabel('Date')
ax.set_ylabel('Furniture Sales')
plt.legend()
plt.show()
```

<font face="Times New Roman" size="3"> The line plot is showing the observed values compared to the rolling forecast predictions. Overall, our forecasts align with the true values very well, showing an upward trend starts from the beginning of the year and captured the seasonality toward the end of the year. </font>

```python
y_forecasted = pred.predicted_mean
y_truth = y['2017-01-01':]
mse = ((y_forecasted - y_truth) ** 2).mean()
print('The Mean Squared Error of our forecasts is {}'.format(round(mse, 2)))
```

```python
print('The Root Mean Squared Error of our forecasts is {}'.format(round(np.sqrt(mse), 2)))
```

<font face="Times New Roman" size="3"> In statistics, the mean squared error (MSE) of an estimator measures the average of the squares of the errors — that is, the average squared difference between the estimated values and what is estimated. The MSE is a measure of the quality of an estimator — it is always non-negative, and the smaller the MSE, the closer we are to finding the line of best fit.

Root Mean Square Error (RMSE) tells us that our model was able to forecast the average daily furniture sales in the test set within 151.84 of the real sales. Our furniture daily sales range from around 400 to over 1200. In my opinion, this is a pretty good model so far. </font>

<font face="Times New Roman" size="3">According to our data, there were way more number of sales from Office Supplies than from Furniture over the years. </font>

```python
furniture = df.loc[df['Category'] == 'Furniture']
office = df.loc[df['Category'] == 'Office Supplies']
furniture.shape, office.shape
```

 <h3><font face="Times New Roman" size="4"> Data Exploration </font> </h3>
 <br>
<font face="Times New Roman" size="3">We are going to compare two categories’ sales in the same time period. This means combine two data frames into one and plot these two categories’ time series into one plot. </font>

```python
cols = ['Row ID', 'Order ID', 'Ship Date', 'Ship Mode', 'Customer ID', 'Customer Name', 'Segment', 'Country', 'City', 'State', 'Postal Code', 'Region', 'Product ID', 'Category', 'Sub-Category', 'Product Name', 'Quantity', 'Discount', 'Profit']
furniture.drop(cols, axis=1, inplace=True)
office.drop(cols, axis=1, inplace=True)
furniture = furniture.sort_values('Order Date')
office = office.sort_values('Order Date')
furniture = furniture.groupby('Order Date')['Sales'].sum().reset_index()
office = office.groupby('Order Date')['Sales'].sum().reset_index()
furniture = furniture.set_index('Order Date')
office = office.set_index('Order Date')
y_furniture = furniture['Sales'].resample('MS').mean()
y_office = office['Sales'].resample('MS').mean()
furniture = pd.DataFrame({'Order Date':y_furniture.index, 'Sales':y_furniture.values})
office = pd.DataFrame({'Order Date': y_office.index, 'Sales': y_office.values})
store = furniture.merge(office, how='inner', on='Order Date')
store.rename(columns={'Sales_x': 'furniture_sales', 'Sales_y': 'office_sales'}, inplace=True)
store.head()
```

```python
plt.figure(figsize=(20, 8))
plt.plot(store['Order Date'], store['furniture_sales'], 'b-', label = 'furniture')
plt.plot(store['Order Date'], store['office_sales'], 'r-', label = 'office supplies')
plt.xlabel('Date'); plt.ylabel('Sales'); plt.title('Sales of Furniture and Office Supplies')
plt.legend();
```

<font face="Times New Roman" size="3">We observe that sales of furniture and office supplies shared a similar seasonal pattern. Early of the year is the off season for both of the two categories. It seems summer time is quiet for office supplies too. in addition, average daily sales for furniture are higher than those of office supplies in most of the months. It is understandable, as the value of furniture should be much higher than those of office supplies. Occasionally, office supplies passed furniture on average daily sales. Let’s find out when was the first time office supplies’ sales surpassed those of furniture’s. </font>

```python
first_date_index = np.min(np.where(store['office_sales'] > store['furniture_sales'])[0])
first_date = store.loc[first_date_index, 'Order Date']
```

```python
print("Office supplies first time produced higher sales than furniture is {}.".format(first_date.date()))
```

# <font face="Times New Roman" size="5">8. Time Series Modeling with Prophet </font>

<font face="Times New Roman" size="3">Released by Facebook in 2017, forecasting tool Prophet is designed for analyzing time-series that display patterns on different time scales such as yearly, weekly and daily. It also has advanced capabilities for modeling the effects of holidays on a time-series and implementing custom changepoints. Therefore, we are using Prophet to get a model up and running. </font>

```python
!pip install prophet

```

```python

import pandas as pd
from prophet import Prophet
```

```python
furniture = furniture.rename(columns={'Order Date': 'ds', 'Sales': 'y'})
furniture_model = Prophet(interval_width=0.95)
furniture_model.fit(furniture)
office = office.rename(columns={'Order Date': 'ds', 'Sales': 'y'})
office_model = Prophet(interval_width=0.95)
office_model.fit(office)
furniture_forecast = furniture_model.make_future_dataframe(periods=36, freq='MS')
furniture_forecast = furniture_model.predict(furniture_forecast)
office_forecast = office_model.make_future_dataframe(periods=36, freq='MS')
office_forecast = office_model.predict(office_forecast)
plt.figure(figsize=(18, 6))
furniture_model.plot(furniture_forecast, xlabel = 'Date', ylabel = 'Sales')
plt.title('Furniture Sales');
```

```python

plt.figure(figsize=(18, 6))
office_model.plot(office_forecast, xlabel = 'Date', ylabel = 'Sales')
plt.title('Office Supplies Sales');
```

<h3><font face="Times New Roman" size="4">Compare Forecasts </font></h3>
<br>
<font face="Times New Roman" size="3">We already have the forecasts for three years for these two categories into the future. We will now join them together to compare their future forecasts. </font>

```python
furniture_names = ['furniture_%s' % column for column in furniture_forecast.columns]
office_names = ['office_%s' % column for column in office_forecast.columns]
merge_furniture_forecast = furniture_forecast.copy()
merge_office_forecast = office_forecast.copy()
merge_furniture_forecast.columns = furniture_names
merge_office_forecast.columns = office_names
forecast = pd.merge(merge_furniture_forecast, merge_office_forecast, how = 'inner', left_on = 'furniture_ds', right_on = 'office_ds')
forecast = forecast.rename(columns={'furniture_ds': 'Date'}).drop('office_ds', axis=1)
forecast.head()
```

<h3><font face="Times New Roman" size="4"> Trend and Forecast Visualization</font></h3>

```python
plt.figure(figsize=(10, 7))
plt.plot(forecast['Date'], forecast['furniture_trend'], 'b-', label='Furniture Trend')
plt.plot(forecast['Date'], forecast['office_trend'], 'r-', label='Office Supplies Trend')
plt.legend()
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Furniture vs. Office Supplies Sales Trend')
plt.show()
```

```python
plt.figure(figsize=(10, 7))
plt.plot(forecast['Date'], forecast['furniture_yhat'], 'b-', label='furniture_yhat')
plt.plot(forecast['Date'], forecast['office_yhat'], 'r-', label='office_yhat')
plt.legend(); 
plt.xlabel('Date'); 
plt.ylabel('Sales')
plt.title('Furniture vs. Office Supplies Estimate');
```
